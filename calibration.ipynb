```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# calibration data
current = np.array([0.16, 0.13, 0.09, 0.06, 0.04, 0.02, 0.01, 0.0])
avg_val = np.array([2128.6, 2125.2, 2119.65, 2115.5076, 2112.746, 2109.7356, 2108.2044, 2106.7608])
```

```python
coeff = np.polyfit(avg_val, current, 1)
print(f'current = {coeff[0]} * x + {coeff[1]}')
```

```python
plt.figure(figsize=(7,4))
plt.plot(avg_val, current)
plt.plot(avg_val, avg_val * coeff[0] + coeff[1])
plt.xlabel('Raw ADC reading')
plt.ylabel('Current through load (A)')
plt.title('Calibration model of current sensors')
plt.legend(['Measured values', 'Linear model'])
plt.savefig('calibration_curve.png')
plt.show()
```
